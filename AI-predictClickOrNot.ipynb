## جایگزین کردن خانه های خالی هر ستون با میانگین همان ستون
    برای این قسمت ابتدا برای هر ستون با کمک تابع ()mean میانگین را به دست آوردیم. با دستور ()dtypes نوع مقادیر آن ستون را به دست می آوریم ومقدار میانگین را با آن تایپ با دستور ()fillna  جایگزین NAN ها میکنیم. این مراحل را برای ستون های دارای NAN و به غیر از ستون هدف که با پیش بینی پر میشود انجام میدهیم.تمام توابع این بخش از کتابخانه pandas هستند. 

```python
featuresWithNan = ["Daily Time Spent on Site", "Age", "Daily Internet Usage"]
for feature in featuresWithNan:
    typeOfFeature = outDataFrame[feature].dtypes
    mean = outDataFrame[feature].mean()
    outDataFrame[feature].fillna(mean.astype(typeOfFeature), inplace = True)
```

```python
outDataFrame.head()
```

    -خروجی این قسمت صحیح است. زیرا همانطور که در بالا میبینیم ستون "Daily Internet Usage" ردیف 1 که در مراحل قبلی خالی بود الان با مقدار میانگین این ستون که در info هم دیده شده بود پر شده است.

```python
outDataFrame.isna().sum()
```

    -دلیل دیگر بر درستی خروجی این قسمت تعداد NAN های هر ستون بعد از این مراحل است که در بالا میبینیم.
    

### بیان دلیل پرکردن خانه های خالی و بیان روشی دیگر و مقایسه با روش میانگین گذاری

## محاسبه میانگین داده ها و مقایسه دو حالت با و بدون حلقه
   

 
 - بدون حلقه (با استفاده از vectorization)
<br>با کمک ()groupby ابتدا وضعیت کلیک را دسته بندی میکنیم و با دستور ()mean میانگین را حساب میکنیم و اطلاعات مربوط به سن را جدا میکنیم.

```python
startTime = time.time()
mean = outDataFrame.groupby("Clicked on Ad").mean()["Age"]
endTime = time.time()
totalTime = endTime - startTime
print(mean)
print("Time with vectorization : %f ms" % (totalTime * 1000))
```

- با حلقه
<br>در اینجا با for تمام داده ها را پیمایش میکنیم و وضعیتشان را تک تک بررسی می کنیم و میانگین را از فرمول آن که مجموع تقسیم بر تعداد است حساب میکنیم.

```python
startTime1 = time.time()
count1 = 0
sum1 = 0
count2 = 0
sum2 = 0
for i in range(outDataFrame["Clicked on Ad"].shape[0]):
    if(outDataFrame.loc[i, "Clicked on Ad"] == 1.0):
        count1 += 1
        sum1 += outDataFrame.loc[i, "Age"]
    if(outDataFrame.loc[i, "Clicked on Ad"] == 0.0):
        count2 += 1
        sum2 += outDataFrame.loc[i, "Age"]
mean1 = sum1 / count1
mean2 = sum2 / count2
endTime1 = time.time()
totalTime1 = endTime1 - startTime1
print("0.0   ",mean2)
print("1.0   ",mean1)
print("Time with Loop : %f ms" % (totalTime1 * 1000))
```

    -همانطور که در بالا میبینیم زمان محاسبات با حلقه بسیار بیشتر از حالت بدون حلقه است. دلیل آن است که در vectorization ما به نحوی در قالب یک دستور خواسته خود را میگوییم و برنامه خود تصمیم میگیرد کجا را میتواند به صورت موازی انجام دهد اما در حلقه مرحله به مرحله میگوییم چه کار کن و برنامه همان قدم ها را باید دنبال کند.

## نمایش توزیع هر ستون از داده
    در این قسمت با کمک فراخوانی تابع hist بر روی هر کدام از ستون ها توزیع داده ها را بر حسب هر یک از مقادیرشان دیدیم. در ادامه این نمودارها آمده است و هرکدام جداگانه تحلیل شده است. hist به صورت پیش فرض روی ستون های عددی اعمال میشود و برای ستونهای غیر عددی میتوان hist را روی value_counts آنها صدا زد. 

```python
hists = outDataFrame.hist(figsize=(25,25))
i = -1
for ax in hists.flatten():
    i += 1 
    title = ax.get_title()
    ax.set_title(str(i)+" :: Distribution of %s" %title,fontweight = "bold")
    ax.set_xlabel(title)
    ax.set_ylabel("Distribution")
```

    در ادامه این پروژه ما میخواهیم در جهتی حرکت کنیم که با تقریب خوبی بتوانیم متغیر هدف را در جاهایی که خالی هستند پیش بینی کنیم. بدین جهت تنها از ویژگی های عددی استفاده میکنیم و بقیه را حذف میکنیم. ویژگی جنسیت را هم که ذاتاً غیر عددی بود و با برچسب گذاری به شکل عدد درآمد را هم حذف میکنیم. برای حذف از تابع ()drop کتابخانه pandas استفاده میکنیم و لیست ستون هایی که میخواهیم حذف شوند را به آن میدهیم و inplace را false میذاریم تا دیتافریم اولیه تغییر نکند.

```python
newDataFrame = outDataFrame.drop(["Ad Topic Line", "City", "Gender", "Country", "Timestamp"], axis = 1, inplace = False)
newDataFrame
```

## نرمال سازی داده ها
    همانطور که گفته شد در ادامه میخواهیم با کمک داده های عددی، هدف را در جاهایی که خالی است پیش بینی کنیم. اگر دقت کنیم میبینیم مقادیر نسبت داده شده به ویژگی های داده شده در یک اسکیل نیستند، مثلاً درآمد هر منطقه از 15000 تا 80000 متغیر است اما سن تنها از 18 تا 63 است. این مسئله کار ما را در آینده از نظر مقایسه و انتخاب بهترین و وابسته ترین ویژگی سخت میکند. بنابراین نیاز است داده ها را نرمال کنیم که همگی به تقریب در یک بازه قرار بگیرد. نرمال سازی راههای متفاوتی دارد اینجا با کم کردن میانگین(محاسبه شده با دستور()mean) از مقادیر و تقسیم کردن بر انحراف معیار(محاسبه شده با دستور()std) داده ها را نرمال میکنیم.

```python
features = ["Daily Time Spent on Site", "Age", "Area Income", "Daily Internet Usage"]
for feature in features:
    mean = newDataFrame[feature].mean()
    std = newDataFrame[feature].std(axis = 0, skipna = True)
    print(feature, "Mean::",mean, "SD::",std)
    newDataFrame[feature] = (newDataFrame[feature] - mean) / std
    newDataFrame[feature]
    print(feature, "Min::",newDataFrame[feature].min(), "Max::",newDataFrame[feature].max())
newDataFrame
```

     -همانطور که در نتایج و داده های بالا میبینیم داده ها نرمال شدند و در بازه ی -3.5 تا 2.8 قرار گرفتند.

## جداسازی داده ها براساس هدف و محاسبه میانگین و انحراف معیار
    در این قسمت برای راحتی مقایسه ویژگی ها در آینده و مقایسه رفتار آنها دو دیتافریم جدید ساختیم؛ یکی مخصوص ویژگی افرادی که روی آگهی کلیک کرده اند(clickedDataFrame) و دیگری مخصوص ویژگی افرادی که روی آگهی کلیک نکرده اند(notClickedDataFrame). همچنین به جهت شناخت بهتر این ویژگی ها برای هر دسته جداگانه میانگین و انحراف معیار را توسط توابع معرفی شده در مرحله قبل محاسبه می کنیم. 

```python
clickedDataFrame = newDataFrame[newDataFrame["Clicked on Ad"] == 1.0]
clickedDataFrame
```

```python
notClickedDataFrame = newDataFrame[newDataFrame["Clicked on Ad"] == 0.0]
notClickedDataFrame
```

```python
features = ["Daily Time Spent on Site", "Age", "Area Income", "Daily Internet Usage"]
featureInfo = {}
for feature in features:
    clickedMean = clickedDataFrame[feature].mean()
    clickedSd = clickedDataFrame[feature].std(axis = 0, skipna = True)
    notClickedMean = notClickedDataFrame[feature].mean()
    notClickedSd = notClickedDataFrame[feature].std(axis = 0, skipna = True)
    featureInfo[feature] = [[clickedMean, clickedSd], [notClickedMean, notClickedSd]]
    print("Mean of %s and CLICKED is::" % feature ,clickedMean)
    print("SD of %s and CLICKED is::" % feature ,clickedSd)
    print("Mean of %s and NOT CLICKED is::" % feature , notClickedMean)
    print("SD of %s and NOT CLICKED is::" % feature , notClickedSd)
    print("-------------------------------------------\n")
```

    -همانطور که در بالا میبینیم میانگین و انحراف معیار هر داده ای در هر دو حالت کلیک کردن یا نکردن محاسبه شده است. برای آنکه در آینده به این اطلاعات نیاز خواهیم داشت این اطلاعات برای هر ویژگی در یک دیکشنری با کلید نام ویژگی ذخیره شده است.

## نمایش نمودار تابع چگالی احتمال (PDF) هر ویژگی
    در این قسمت برای هر حالت کلیک کردن و نکردن به ازای هر ویژگی تابع چگالی آن را رسم کردیم. چگالی احتمال به کمک تابع norm.pdf که متعلق به کتابخانه scipy.stats است به راحتی محاسبه میشود. این تابع سه ورودی میگیرد: 1- متغیر 2- میانگین متغیر 3- انحراف از معیار متغیر. میانگین و انحراف معیار را از دیکشنری حاصل مرحله قبل استخراج کردیم. همچنین برای مقایسه راحت تر برای هر ویژگی هر دو حالت کلیک و عدم کلیک در یک نمودار و هم با نقطه و هم خط رسم شده. در ادامه این نمودارها را میبینید. 

```python
features = ["Daily Time Spent on Site", "Age", "Area Income", "Daily Internet Usage"]
i = -1
for feature in features:
    i += 1
    x1 = clickedDataFrame[feature]
    y1 = norm.pdf(x1, featureInfo[feature][0][0], featureInfo[feature][0][1])
    x2 = notClickedDataFrame[feature]
    y2 = norm.pdf(x2, featureInfo[feature][1][0], featureInfo[feature][1][1])
    mplot.figure(figsize=(15,5))
    mplot.subplot(1, 2, 1)
    mplot.plot(x1, y1)
    mplot.plot(x2, y2)
    mplot.legend(["clicked","not clicked"])
    mplot.title(str(i)+" :: PDF of %s - PLOT" % feature,fontweight = "bold")
    mplot.xlabel("Normalized %s" % feature)
    mplot.ylabel("PDF")
    i += 1
    mplot.subplot(1, 2, 2)
    mplot.scatter(x1, y1)
    mplot.scatter(x2, y2)
    mplot.legend(["clicked","not clicked"])
    mplot.title(str(i)+" :: PDF of %s - SCATTER" % feature,fontweight = "bold")
    mplot.xlabel("Normalized %s" % feature)
    mplot.ylabel("PDF")
    mplot.show()
```

    - برای تحلیل نمودارهای حاصل ابتدا بهتر است مفهوم تابع چگالی احتمال را مرور کنیم. تابع چگالی احتمال یعنی برای هر متغیری به اندازه ی مساحت زیر نمودار آن ناحیه، احتمال دارد متغیر در آن ناحیه قرار بگیرد. با توجه به این تعریف ما دنبال ویژگی هستیم که راحت بتوان برای دو حالت نمودار ها و در اصل احتمال رخ دادن را تمییز داد. یعنی به دنبال نمودارهایی هستیم که کمترین میزان اشتراک در دو حالت را داشته باشند و بتوانیم بگوییم از جایی به بعد به احتمال بیشتر کاربر کلیک کرده یا نکرده. 
     نمودار 0 و 1: این دو نمودار چگالی احتمال کلیک و عدم کلیک ویژگی زمان سپری شده در سایت است. 
     نمودار 2 و 3: این دو نمودار چگالی احتمال کلیک و عدم کلیک ویژگی سن است. 
     نمودار 4 و 5: این دو نمودار چگالی احتمال کلیک و عدم کلیک ویژگی درآمد منطقه است. 
     نمودار 6 و 7: این دو نمودار چگالی احتمال کلیک و عدم کلیک ویژگی مصرف روزانه اینترنت است.
     در نمودار های سن و درآمد منطقه مقدار اشتراک چشم گیر و زیادی به چشم میخورد و حذف میشوند ولی برای مصرف روزانه اینترنت و زمان سپری شده در سایت به مقایسه دقیق تری نیاز داریم. 

```python
x = clickedDataFrame["Daily Internet Usage"]
x1 = clickedDataFrame["Daily Time Spent on Site"]
x2 = notClickedDataFrame["Daily Internet Usage"]
x3 = notClickedDataFrame["Daily Time Spent on Site"]
clickUsage =norm.pdf(clickedDataFrame["Daily Internet Usage"], featureInfo["Daily Internet Usage"][0][0], featureInfo["Daily Internet Usage"][0][1])
noClickUsage = norm.pdf(notClickedDataFrame["Daily Internet Usage"], featureInfo["Daily Internet Usage"][1][0], featureInfo["Daily Internet Usage"][1][1])
clickTime = norm.pdf(clickedDataFrame["Daily Time Spent on Site"], featureInfo["Daily Time Spent on Site"][0][0], featureInfo["Daily Time Spent on Site"][0][1])
noClickTime =norm.pdf(notClickedDataFrame["Daily Time Spent on Site"], featureInfo["Daily Time Spent on Site"][1][0], featureInfo["Daily Time Spent on Site"][1][1])
mplot.figure(figsize=(15,5))
mplot.subplot(1, 2, 1)
mplot.scatter(x, clickUsage)
mplot.scatter(x1, clickTime)
mplot.legend(["Usage","Time"])
mplot.title("CLICK",fontweight = "bold")
mplot.xlabel("Normalized data")
mplot.ylabel("PDF")
mplot.subplot(1, 2, 2)
mplot.scatter(x2, noClickUsage)
mplot.scatter(x3, noClickTime)
mplot.legend(["Usage","Time"])
mplot.title("NOT CLICK",fontweight = "bold")
mplot.xlabel("Normalized data")
mplot.ylabel("PDF")
mplot.show()
```

- برای مقایسه بهتر کلیک و عدم کلیک دو ویژگی را جدا کردیم و کلیک برای دو ویژگی را باهم و عدم کلیک آن دو را هم با هم میکشیم. در هردو عدم کلیک با کلیک از سمت چپشان تداخل دارد که میبینیم باهم یکسانند اما کلیک ها که با عدم کلیک از سمت راست تداخل دارند تفاوت کوچکی دارند که نشان میدهد ناحیه اشتراک **میزان استفاده روزانه کاربر از معیار مناسبی برای پیش بینی ویژگی هدف است.** 

## پیش بینی 
    -برای پیش بینی باید نقطه تلاقی دو حالت کلیک و عدم کلیک نمودار میزان استفاده روزانه کاربر را بیابیم . سپس با توجه مقدار نرمال ویژگی میزان استفاده روزانه کاربر و جایگاه آن نسبت به نقطه تلاقی تصمیم میگیریم هدف چه باشد.
    در اینجا برای نقطه تلاقی از حدس استفاده شده و برای هر حدس در ناحیه مثبت و منفی شکل رسم شده. سپس با زوم در تصویر و بررسی دیتافریم در ناحیه ها سعی شده دقیق ترین مقدار پیدا شود.

```python
feature = "Daily Internet Usage"
num = [0.02, 0.025, 0.035, 0.05]
for i in num:
    x1 = clickedDataFrame[feature]
    y1 = norm.pdf(x1, featureInfo[feature][0][0], featureInfo[feature][0][1])
    x2 = notClickedDataFrame[feature]
    y2 = norm.pdf(x2, featureInfo[feature][1][0], featureInfo[feature][1][1])
    mplot.figure(figsize=(15,5))
    mplot.subplot(1, 2, 1)
    mplot.plot(x1, y1,'.')
    mplot.plot(x2, y2,'.')
    mplot.axvline(x=i)
    mplot.legend(["clicked","not clicked"])
    mplot.title("PDF of %s  " % feature + str(i),fontweight = "bold")
    mplot.xlabel("Normalized %s" % feature)
    mplot.ylabel("PDF")
    mplot.subplot(1, 2, 2)
    mplot.plot(x1, y1, '.')
    mplot.plot(x2, y2, '.')
    mplot.axvline(x=-i)
    mplot.legend(["clicked","not clicked"])
    mplot.title("PDF of %s  " % feature + str(-i),fontweight = "bold")
    mplot.xlabel("Normalized %s" % feature)
    mplot.ylabel("PDF")
```

## پر کردن خانه های خالی هدف
    - بدین جهت ابتدا از روی دیتافریم با داده های نرمال یک دیتافریم جدید که مخصوص خانه های خالی ستون هدف بود ساختم. سپس در این دیتافریم داده هایی که مقدار ویژگی مصرف روزانه اینترنتشان کمتر از نقطه تلاقی بود را به عنوان داده های کلیک کرده در یک دیتافریم جدید ریختم و با ترکیب اندیس دیتافریم خالی و دیتافریم کلیک ها (دستور concat) یک دیتافریم ساختم که در این خانه هایی که درهدف خالی موندند خانه های عدم کلیک هستند. سپس با دستور to_csv دیتافریم نهایی را در یک فایل csv با نام out.csv ذخیره کردم.
